```python
import findspark
findspark.init()

from pyspark import SparkContext
from pyspark import SparkConf
```

```python
from pyspark.sql import SparkSession
import pandas as pd

spark = SparkSession.builder.getOrCreate()
```

```python
#csv 파일 읽기
#header=True
#inferSchema=True : 각 데이터 타입 추론해 자동 조정 안할 시 모든 value는 String으로
df_pyspark = spark.read.csv('./sample_dataset.csv', header=True, inferSchema=True)
df_pyspark
```

```python
#스키마 및 컬럼 확인
df_pyspark.printSchema()
df_pyspark.columns
```

```python
#n개의 record 데이터
df_pyspark.show(1, vertical=True)
```

```python
#특정 열에 대한 분석 
df_pyspark.select("loan_amnt").describe().show()
```

```python
#모든 행에 대한 정보
#메모리 초과 에러 발생 가능
df_pyspark.collect()

#대안
#특정 행에 대한 정보
df_pyspark.take(1)
```

```python
#record의 모든 column이 null 값인 것 제거
#df_pyspark.na.drop(how="all").show()
#record의 어떤 하나의 column이라도 null 값인 것 제거
df_pyspark.na.drop(how="any").show()
#record의 subset 컬럼이 null인 것 제거
df_pyspark.na.drop(how="any", subset=['emp_title']).show()
```

```python
#column 접근
df_pyspark.emp_title
```

```python
#filter
df_pyspark.filter(df_pyspark.id == 1).show()
```

```python
#중간값 나아가 Q1, Q2, Q3 
df_pyspark.approxQuantile("loan_amnt", [0.25], 0.25)
```
